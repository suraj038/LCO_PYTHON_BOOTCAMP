```python
import requests
import bs4
res=requests.get('https://www.flipkart.com/search?q=tricycles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
soup=bs4.BeautifulSoup(res.text,'lxml')
scrap=soup.findAll('a', {'class': '_2cLu-l'})
scrap1=soup.findAll('div', {'class': '_1vC4OE'})
dct={}
for i,j in zip(scrap,scrap1):
    dct[i.text]=j.text
    print(dct)
```
